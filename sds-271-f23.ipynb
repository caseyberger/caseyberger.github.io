# Labs

You can find all labs in Moodle, or in this [Google Drive Folder](https://drive.google.com/drive/folders/1BrjKefesMYX1B9Gv8545PNQnDzq5kt4d?usp=drive_link/).

## Lab 1

[Lab 1](SDS271/F23/Lab1_clean.qmd) is all about getting started with JupyterHub and reviewing your previous Python experience.

You can find the .ipynb for this lab in Moodle, or at [this link](https://colab.research.google.com/drive/1w4aK3JO1ifkciSFsfkBdUUlgqYBZWBN2).

[Lab 1 solutions](SDS271/F23/Lab1_solutions.qmd)

## Lab 2

[Lab 2](SDS271/F23/Lab2_clean.qmd) gives you a chance to dig deeper into Python; learn how to import and use Python packages like random, matplotlib, and numpy; and apply your previous statistical knowledge to explore the dangers of making uninformed wagers.

[Lab 2 Solutions](SDS271/F23/Lab2_solutions.qmd)

## Lab 3

[Lab 3](SDS271/F23/Lab3_clean.qmd) introduces the data science package Pandas, which allows you to organize, filter, sort, and do simple statistics on large datasets.

## Lab 4

[Lab 4](SDS271/F23/Lab4_clean.qmd) digs deeper into pandas and matplotlib using data from nuclear physics. It also introduces the seaborn visualization package.
